# **1. 문자열 거꾸로 출력하기**

입력된 문자를 재귀함수를 이용해 거꾸로 출력하는 프로그램을 작성하여라.

```python
def printReverse(str):
  if str == "":
    return str
  else:
    return str[1:] + str[0] #문자열의 길이가 1보다 길면 맨 앞의 문자를 뒤로 보내고 나머지 문자열에 대해 재귀적 시행함
```

```python
printReverse("Reverse")
```

# **2. 노드 짝지어 교환하기**

주어진 링크드 리스트에서 이웃한 노드끼리 두 개씩 교환하고 헤드를 반환하는 프로그램을 작성하여라.

```python
#링크드 리스트 클래스를 직접 선언하여 만든 함수.
#링크드 리스트를 구성하는 노드 클래스 선언문
class Node:
  def __init__(self, data, pointer = None):
    self.data = data
    self.pointer = pointer
#링크드 리스트 클래스 선언문
class LinkedList(object):
  def __init__(self):
      self.head = Node(None)
      self.size = 0
#노드의 인덱스 반환
  def get(self, dex):
    if(dex >= self.size):
      print("Index Error")
      return Node
    if(dex == 0):
      return self.head
    else:
      node = self.head
      for _ in range(dex):
        node = node.pointer
      return node
#리스트 끝에 새 노드 추가
  def append(self, data):
    if(self.size == 0):
      self.head = Node(data)
      self.size += 1
    else:
      node = self.head
      while(node.pointer != None):
        node = node.pointer
      
      new_node = Node(data)
      node.pointer = new_node
      self.size += 1
#인덱스에 해당하는 노드 삭제
  def delete(self, dex):
    if(self.size == 0):
      return
    elif(dex >= self.size):
      return
    elif(dex == 0):
      self.head = self.head.pointer
      self.size -= 1
    else:
      node = self.get(dex-1)
      node.pointer = node.pointer.pointer
      self.size -= 1
#인덱스 위치에 노드 삽입하기
  def insert(self, data, dex):
    if(self.size == 0):
      self.head = Node(data)
      self.size += 1
    elif(dex == 0):
      self.head = Node(data, self.head)
      self.size += 1
    else:
      node = self.get(dex - 1)
      if(node == None):
        return
      newNode = Node(data)
      nextNode = node.pointer
      node.pointer = newNode
      newNode.pointer = nextNode
      self.size += 1
#리스트 출력하기
  def print_linked_list(self):
    list = []
    node = self.head
    while node:
      list.append(node.data)
      node = node.pointer
    for _ in range(len(list)):
      print(list[_], end=" ")
    print()
#링크드 리스트 내 순서 바꾸는 함수
def swap_Nodes_pair(list):
  if(list.size <= 2):
    if(list.size == 2):
      firstnode = list.get(0).data
      secondnode = list.get(1).data
      list.delete(1)
      list.insert(secondnode, 0) #길이가 2이면 두번째 노드 삭제 후 첫번째에 삽입한 다음 반환
    return
  else:
    firstnode = list.get(0).data
    secondnode = list.get(1).data
    list.delete(0)
    list.delete(0)
    swap_Nodes_pair(list)
    list.insert(firstnode, 0)
    list.insert(secondnode, 0) #길이가 2보다 길면 앞의 두 노드 삭제한 리스트에 대해 재귀적 시행한 다음 앞에 순서를 바꾸어 노드를 다시 넣음
```

```python
numberlist = LinkedList()
for _ in range(13):
  numberlist.append(_)
numberlist.print_linked_list()
swap_Nodes_pair(numberlist)
numberlist.print_linked_list()
```

# **3. 파스칼의 삼각형**

왼쪽변과 오른쪽변이 모두 1로 이루어져 있고 나머지 칸은 위의 이웃한 두 개의 수를 더한 값을 나타내는 삼각형을 출력하라.

```python
def Pascal(number1, number2):
  cache = [[None] * number2 for _ in range(number1)] #파스칼 수가 들어갈 배열 초기화
  def Pascal_triangle(number1, number2):
    if(cache[number1-1][number2-1] != None):
      return cache[number1-1][number2-1] #해당 함수값이 cache 안에 있으면 cache 값을 반환함

    if(number1 == number2 or number2 == 1):
      result = 1
    else:
      result = Pascal_triangle(number1 - 1, number2 - 1) + Pascal_triangle(number1 - 1, number2)
    
    cache[number1-1][number2-1] = result #계산 결과 cache에 저장 후 반환함
    return result
  return Pascal_triangle(number1, number2)

def Make_Pascal_triangle(number):
  for _ in range(1, number+1):
    for __ in range(1, _+1):
      print(Pascal(_, __), end=" ")
    print()
```

```python
Make_Pascal_triangle(10)
```

# **4. 파스칼의 삼각형 가로줄 출력하기**

파스칼의 삼각형 중에서 rowindex에 해당하는 가로줄을 모두 출력하여라.

```python
#파스칼 삼각형 만드는 과정은 3과 동일함
def Pascal(number1, number2):
  cache = [[None] * number2 for _ in range(number1)]
  def Pascal_triangle(number1, number2):
    if(cache[number1-1][number2-1] != None):
      return cache[number1-1][number2-1]

    if(number1 == number2 or number2 == 1):
      result = 1
    else:
      result = Pascal_triangle(number1 - 1, number2 - 1) + Pascal_triangle(number1 - 1, number2)
    
    cache[number1-1][number2-1] = result
    return result
  return Pascal_triangle(number1, number2)

def Make_Pascal_line(rowindex):
  list = []
  for _ in range(1, rowindex+1):
    list.append(Pascal(rowindex, _))
  print(list)
```

```python
Make_Pascal_line(10)
```

# **5. 계단 오르기**

n개의 계단을 걸어올라 맨 위에 도달하고자 한다. 한 번에 1계단 또는 2계단씩만 오를 수 있을 때 몇 가지의 방법으로 계단을 오를 수 있는가?

```python
#피보나치 수열의 원리를 이용해 재귀적으로 구현함
def climbing_stairs(N):
  if(N == 1):
    return 1
  elif(N == 2):
    return 2
  else:
    return climbing_stairs(N - 1) + climbing_stairs(N - 2)
```

```python
climbing_stairs(13)
```

# **6. 거듭제곱 계산하기**

pow(x, n) = x ^ n을 recursion을 이용해 계산하여라.

```python
#지수의 값이 0보다 크면 1 작은 값에 대한 결과에 밑을 곱하고 작으면 밑을 나눔. 0일시 1 반환
def POW_recursive(x, n):
  if(n > 0):
    return POW_recursive(x, n - 1) * x
  elif(n < 0):
    return POW_recursive(x, n + 1) / x
  else:
    return 1
```

```python
print(POW_recursive(2, 10))
```

```python
print(POW_recursive(3.7, 8))
```

```python
print(POW_recursive(4.2, -6))
```

# **7. 순열 출력하기**

여러 개의 정수로 이루어진 배열을 입력하면 정수들을 조합해 나올 수 있는 모든 순열을 출력하는 프로그램을 작성하여라. (중복된 순서쌍은 제외한다.)

```python
def Make_Permutation(list):
  if(len(list) == 1):
    return [list] #전체 배열의 길이가 1이면 해당 원소 리스트 형식으로 반환함
  cache = [] #결과 저장할 배열 초기화
  for _ in list:
    newlist = [number for number in list if number != _] #배열 원소 중 반복 인덱스에 해당하는 원소를 제외한 배열 만듬
    numberlist = Make_Permutation(newlist) #사용한 원소를 제외한 배열에 대해 재귀적으로 시행함
    for __ in numberlist:
      cache.append([_] + __) #반복 인덱스 원소 값 이후 재귀적 시행한 배열의 결과값을 차례로 넣음
  return cache
```

```python
numberlist = [2, 3, 5, 7]
print(Make_Permutation(numberlist))
```
