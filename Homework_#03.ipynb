# **1. 도둑 집 털기**

거리에 있는 집들은 일정 금액의 돈이 숨겨져 있다. 인접한 집끼리는 보안 시스템이 연결되어 있어 같은 날 이웃한 두 집에 침입하면 자동으로 경찰에 연락이 간다. 경찰에 들키지 않고 오늘 밤 도둑질 할 수 있는 최대 금액은 얼마인가?

```python
def RobberMoney(houselist):
  if(len(houselist) == 1): #집이 하나일 경우 해당 집의 금액 반환함
    return houselist[0]
  elif(len(houselist) == 2): #집이 두 개일 경우 더 많은 돈이 있는 집의 금액 반환함
    return max(houselist[0], houselist[1])
  elif(len(houselist) % 2 == 0): #집이 짝수 개일 경우 맨 앞의 집을 선택하고 건너 집부터 탐색하는 방법과 맨 뒤의 집을 선택하고 건너 집부터 탐색하는 방법 중 최대 이익을 얻는 결과를 반환함
    return max(houselist[0] + RobberMoney(houselist[2:]), houselist[len(houselist) - 1] + RobberMoney(houselist[:len(houselist) - 2]))
  elif(len(houselist) % 2 == 1): #집이 홀수 개일 경우 짝수 개의 경우에 덧붙여 맨 앞과 맨 뒤의 집을 제외한 집을 탐색하는 방법을 포함하여 최대 이익을 얻는 결과를 반환함
    return max(houselist[0] + RobberMoney(houselist[2:]), RobberMoney(houselist[1:len(houselist) - 1]), houselist[len(houselist) - 1] + RobberMoney(houselist[:len(houselist) - 2]))
```

```python
houselist = [2, 7, 9, 3, 1]
print("Maximum Output: ", end="")
print(RobberMoney(houselist))
```

# **2. 최소 비용으로 계단 오르기**

계단을 오를 때마다 특정 비용이 발생한다고 한다. 계단은 한 계단 혹은 두 계단씩 오를 수 있다고 할 때, 꼭대기 층에 도달하는 최소 비용을 구하시오. (인덱스는 0 또는 1부터 시작)

```python
def TotalStairCost(cost):
  if(len(cost) == 1): #계단이 1개일 경우 첫번째 계단에 대한 비용을 반환하고 2개일 경우 두번째 계단을 오르는 비용을 반환함
    return cost[0]
  elif(len(cost) == 2):
    return cost[1]
  else: #계단이 3개 이상이면 1칸/2칸 우선 오른 후 나머지 계단을 오르는 방법과 마지막에 1칸/2칸 오르는 방법 중 비용이 최소가 되는 결과를 반환함
    return min(cost[0] + TotalStairCost(cost[1:]), cost[1] + TotalStairCost(cost[2:]), cost[len(cost) - 2] + TotalStairCost(cost[:len(cost) - 2]), cost[len(cost) - 1] + TotalStairCost(cost[:len(cost) - 1]))
```

```python
cost = [1,100,1,1,1,100,1,1,100,1]
print("최소 비용: ", end="")
print(TotalStairCost(cost))
```

# **3. 레벤슈타인 거리 구하기**

철자가 틀린 문자열이 주어지면 맞춤법 검사기는 해당 문자열과 가장 가까운 단어를 반환한다. 이때 문자열을 변환하는데 필요한 최소 편집 횟수를 레벤슈타인 거리라고 한다.
문자열 2개가 입력되었을 때 첫번째 문자열에서 두번째 문자열로 변환하는 데 필요한 레벤슈타인 거리를 구하여라.

```python
def WordDistance(word1, word2):
  dex1 = len(word1)
  dex2 = len(word2)
  cache = [[None] * (dex2 + 1) for _ in range(dex1 + 1)] #두 단어의 길이에 맞는 배열을 초기화
  def LevenshteinDistance(word1, word2, number1, number2):
    if(cache[number1][number2] != None):
      return cache[number1][number2] #배열 안에 값이 존재할 경우 배열 값 반환함
    #가로줄과 세로줄이 시작되는 곳일 경우 해당 단어의 인덱스 값을 반환함
    if(number1 == 0):
      result = number2
    elif(number2 == 0):
      result = number1
    else:
      if(word1[number1 - 1] == word2[number2 - 1]): #해당 알파벳끼리 동일하면 왼쪽 위에 저장된 값을 불러옴
        result = LevenshteinDistance(word1, word2, number1 - 1, number2 - 1)
      else: #알파벳이 서로 다르면 왼쪽 위, 위 그리고 왼쪽의 값 중 최솟값에 1을 더한 값을 저장함
        choice1 = LevenshteinDistance(word1, word2, number1 - 1, number2 - 1)
        choice2 = LevenshteinDistance(word1, word2, number1, number2 - 1)
        choice3 = LevenshteinDistance(word1, word2, number1 - 1, number2)
        result = min(choice1, choice2, choice3) + 1
    cache[number1][number2] = result
    return result
  return LevenshteinDistance(word1, word2, dex1, dex2)
```

```python
Word1 = "Carthorse"
Word2 = "Orchestra"
print("레벤슈타인 거리: ", end="")
print(WordDistance(Word1, Word2))
#철자를 비교해서 나오는 숫자를 확인하기 위해 출력문 사용함
for _ in range(1, len(Word1) + 1):
  for __ in range(1, len(Word2) + 1):
    print(WordDistance(Word1[:_], Word2[:__]), end=" ")
  print()
```

```python
Word1 = "Saturday"
Word2 = "Sunday"
print("레벤슈타인 거리: ", end="")
print(WordDistance(Word1, Word2))
#첫번째 단어가 세로줄이고 두번째 단어가 가로줄로 들어감
for _ in range(1, len(Word1) + 1):
  for __ in range(1, len(Word2) + 1):
    print(WordDistance(Word1[:_], Word2[:__]), end=" ")
  print()
```
